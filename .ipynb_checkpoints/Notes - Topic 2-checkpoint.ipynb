```python
def count_char_types(s):
    '''counts types of characters in string'''
    numbers = sum(c.isdigit() for c in s)
    lower = sum(c.islower() for c in s)
    upper = sum(c.isupper() for c in s)
    spaces  = sum(c.isspace() for c in s)
    others  = len(s) - numbers - lower - upper - spaces
    return pd.Series([numbers, lower, upper, spaces, others])
```

```python
import nltk
from nltk import word_tokenize #pos_tag,
from nltk.corpus import stopwords
import regex as re
import webcolors

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
descriptions = revenue_codes[['description']].copy()
```

```python
color_list = list(webcolors.CSS3_NAMES_TO_HEX.keys())
```

```python
'red' in 'spots on red bookcover tape'
```

```python
def extract_color(string):
    colors = []
    string = string.lower()
    for i in color_list:
        if i in string:
            string = string.replace(i, "")
            colors.append(i)
    return string, colors
```

```python
def extract_non_text(string):
    string = string.lower()
    extract = "".join(re.findall(r'[^a-zA-Z\s]',string))
    string = "".join(re.findall(r'[a-zA-Z\s]',string))
    return string, extract
```

```python
def extract_stopwords(string):
    colors = []
    string = string.lower()
    for i in color_list:
        if i in string:
            string = string.replace(i, "")
            colors.append(i)
    return string, colors
```

```python
descriptions['clean'], descriptions['color']= zip(*descriptions['description'].apply(extract_color))
```

```python
descriptions['color'].value_counts()
```

```python
descriptions['clean'], descriptions['num_special_char'] = zip(*descriptions['clean'].
                                                              apply(extract_non_text))
```

```python
cvec = CountVectorizer()
cvec.fit(X)
X = cvec.transform(X)
```

```python
X = pd.DataFrame(X.todense(),
             columns=cvec.get_feature_names_out()
            )
```
