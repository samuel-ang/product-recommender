```python
def count_char_types(s):
    '''counts types of characters in string'''
    s=str(s)
    numbers = sum(c.isdigit() for c in s)
    lower = sum(c.islower() for c in s)
    upper = sum(c.isupper() for c in s)
    spaces  = sum(c.isspace() for c in s)
    others  = len(s) - numbers - lower - upper - spaces
    return pd.Series([numbers, lower, upper, spaces, others])
```

Identify latest invoice by customer

```python
# Invoice date combinations
invoice_date = sales_data[['customerid',
                           'invoiceno',
                           'invoicedate']].value_counts().reset_index().drop(columns=0)
```

```python
# Check if theres only one date per invoice
multiple_date = invoice_date['invoiceno'].value_counts().loc[lambda x: x>1]
```

```python
multiple_date.index
```

```python
for i in multiple_date.index:
    select_invoice = invoice_date[invoice_date['invoiceno']==i]
    difference = select_invoice['invoicedate'].iloc[1] - select_invoice['invoicedate'].iloc[0]
    print(i, difference.total_seconds() / 60)
```

The datetimes of invoices with two datetimes are all one minute apart. We would select the later datetime to use.

```python
latest_datetime = invoice_date.groupby(['invoiceno'])['invoicedate'].apply(
                                (lambda x : max(x))).reset_index()
```

```python
latest_datetime.info()
```

```python
invoice_cust.info()
```

```python
invoice_cust = invoice_cust.merge(latest_datetime, on='invoiceno')
```

```python
invoice_cust_excl_return = invoice_cust[~invoice_cust['invoiceno'].str.contains('C')]
```

```python
latest_sales_invoices = invoice_cust_excl_return.loc[invoice_cust_excl_return
                                                     .groupby('customerid').invoicedate.idxmax()]
```
